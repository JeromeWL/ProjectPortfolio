```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
from math import sin, cos
import time
```

```python
#Function defintion
def f(x):
    return sin(x)

#Derivative of that function
def fd(x):
    return cos(x)
```

```python
#Parameters:
#Starting X value,
#Learning rate or how large the step size between values 
#Ammount of iterations
#Minimal tolerance, or how close teh value needs to be to the minimum before ending early.

def gradientDescent(startingX, lr, iter, minTol):
    x = startingX
    history = [x]
    timeHistory = []
    
    for i in range(iter):
        start_time = time.time()  # Record start time
        
        gradient = fd(x)
        newval = x - lr * gradient
        
        # Tolerance check
        if abs(newval - x) < minTol:
            # print(f'Min tolerance reached at {i} iterations')
            break
        
        x = newval
        history.append(x)
        
        end_time = time.time()  # Record end time
        elapsed_time = end_time - start_time  # Calculate the elapsed time for this iteration
        timeHistory.append(elapsed_time)  # Append the elapsed time to the list
    
    # Optionally, print the average time for an iteration
    print(f'Average time per iteration: {sum(timeHistory)/len(timeHistory):.20f} seconds')
    print(f'Total time for {iter} iterations: {sum(timeHistory):.20f} seconds')

    
    return x, history
```

```python
def visualize_gradient_descent(startingX, lr, iter, minTol):
    final_x, history = gradientDescent(startingX, lr, iter, minTol)
    
    #Graph range values
    x_range = np.linspace(min(history)-10, max(history)+10, 500)
    y_range = [f(x) for x in x_range]
    
    #Basic plot definitions
    plt.figure(figsize=(12, 6))
    
    #Plotting the function line
    plt.plot(x_range, y_range, 'b-',)
    
    #Function point definitions
    plt.scatter(history, [f(x) for x in history], color='gray', label='Gradient Descent Path')
    plt.scatter(startingX, f(startingX), color='green', s=100, label='Starting Point')
    plt.scatter(final_x, f(final_x), color='red', s=100, label='Final Point')
    
    #Misc
    plt.title('Gradient Descent Optimization of a function')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    #Additional info thats not bad to have
    print(f'Starting X: {startingX}')
    print(f'Final X: {final_x}')
    print(f'Final y: {f(final_x)}')
    print(f'Number of iterations: {len(history)}')
```

```python
visualize_gradient_descent(startingX=1, lr=0.25, iter=500, minTol=0.001)
```

```python
visualize_gradient_descent(startingX=1.7, lr=0.15, iter=500, minTol=0.01)
```
